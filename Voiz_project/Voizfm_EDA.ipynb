### Recency

#### 1.CALCULATE Recency

```python
# Recency: listen_minute per week: thời lượng nghe trung bình trong 1 tuần (tính theo đơn vị phút)
```

```python
## get the max purchase date for each customer and create a dataframe with it
## take our observation point as the max invoice date in our dataset
order_summary_df['Recency'] = (order_summary_df['max_purchase_date'].max() - order_summary_df['max_purchase_date']).dt.days

print('MaxPurchaseDate: '+str(order_summary_df['max_purchase_date'].max()))
order_summary_df.head(3)

# users with 0_Recency are the most active ones
```

```python
## plot a recency histogram
sns.histplot(data=order_summary_df,x='Recency')
```

```python
## describe recency data
order_summary_df['Recency'].describe()
```

#### 2.DETERMINE K

```python
# for Ebowl
def find_sse(df, max_k):
    sse = {}
    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k, max_iter=1000)
        kmeans.fit(df)
        df['clusters'] = kmeans.labels_
        sse[k] = kmeans.inertia_
    return sse
   
## function for ordering cluster numbers
def order_cluster(cluster_field_name, target_field_name, df, ascending):
    new_cluster_field_name = 'new_' + cluster_field_name
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final
```

```python
## Elbow Method to determine K

sse = find_sse(order_summary_df.loc[:,['Recency']], max_k = 10) 

plt.figure()
plt.plot(list(sse.keys()),list(sse.values()))
plt.xlabel("Number of clusters")
plt.show() 

# Optimal K=3
## Choose the k of Recency_clusters = 3
## 1st: 'active', 2st: 'sleep', 3rd: 'hibernate'
```

#### 3.CLUSTERING

```python
## k-means = 4
kmeans = KMeans(n_clusters=4)
kmeans.fit(order_summary_df[['Recency']])
order_summary_df['RecencyCluster'] = kmeans.predict(order_summary_df[['Recency']])

order_summary_df = order_cluster('RecencyCluster', 'Recency',order_summary_df,False)

## see details of each cluster
order_summary_df.groupby('RecencyCluster')['Recency'].describe()

## ClusterRecency: 0: ít active nhất(hibernate): active trong vòng 10 tháng,
# 1: active trong vòng 7 tháng
# 2: active vừa vừa (sleep): active trong vòng 3.5 tháng, 
# 3: the most active (active) - active trong vòng 1 tháng
```
